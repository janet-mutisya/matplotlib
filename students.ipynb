```python
#count students per year
year_counts = df['year'].value_counts().sort_index()
import matplotlib.pyplot as plt
#plot the linechart
plt.plot(year_counts.index, year_counts.values, marker='o', linestyle='-', color='b', label='numbe of students')
plt.title('students_Table')
plt.xlabel('year')
plt.ylabel('no. of students')
plt.xticks(year_count.index)

plt.legend()
plt.tight_layout()
plt.show()

```

```python
#group by year and calculate the avarage mean
avg_age_per_year = df.groupby("year")['age'].mean()
#plot the bar chat
plt.bar(avg_age_per_year.index, avg_age_per_year.values, color='blue')
plt.title('students per year')
plt.xlabel('year')
plt.ylabel('age')
plt.xticks(avg_age_per_year.index)

plt.tight_layout()

plt.show()
```
