```python
import matplotlib.pyplot as plt
import pandas as pd
#load the data
df = pd.read_csv('owid-covid-data .csv', usecols=['location', 'date', 'total_cases'])

#up to  date data total cases
latest_data = df.sort_values('date').groupby('location').tail(1)
#clean the data
latest_data = latest_data.dropna(subset=['total_cases'])
latest_data = latest_data[~latest_data['location'].isin(['World', 'International'])]
#get the top ten countries
top_countries = latest_data.nlargest(10, 'total_cases')
#plotting
plt.figure(figsize=(12, 6))
plt.bar(top_countries['location'], top_countries['total_cases'], color='skyblue')
plt.xticks(rotation=45)
plt.title('Top 10 Countries  affected by COVID-19 ')
plt.ylabel('Total Cases')
plt.xlabel('Country')
plt.tight_layout()
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('owid-covid-data.csv', usecols=['location', 'date', 'people_vaccinated'])

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# Select countries of interest
countries = ['United States', 'India', 'Germany', 'Brazil']
vacc_data = df[df['location'].isin(countries)]

# Plot line chart
plt.figure(figsize=(12, 6))
for country in countries:
    country_data = vacc_data[vacc_data['location'] == country]
    plt.plot(country_data['date'], country_data['people_vaccinated'], label=country)

plt.title('Cumulative People Vaccinated Over Time')
plt.xlabel('Date')
plt.ylabel('People Vaccinated')
plt.legend()
plt.tight_layout()
plt.show()
```
